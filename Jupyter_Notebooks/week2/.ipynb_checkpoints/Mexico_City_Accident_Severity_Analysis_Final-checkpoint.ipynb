## Introduction
Mexico-City is considered as one of the most problematic and conflict cities in the world. Everyday, it is reported lot of accidents. To improve the system, the government decide to include help-buttons in the electric poles to help people report incidents. ### Reading the database
The database has a length of 1,355,257 with all the information generated by means of the C4 system. Uploading the information to GitHub is not possible due to its size. For that reason, all the information was splitted by means of the year and uploading to github.
Furthermore, there is another problem how to collect all the information previously stored and used to perform the data analysis. The following part of the code is to read the database named *incidentes-viales-c5.csv*. Nevertheless, this file cannot be uploaded to GitHub due to its size. To solve this problem, a *dataSet_incidentes-viales-c5.zip* was uploaded with all the databases splited by year. Furthemore, to obtain the full database can download from https://datos.cdmx.gob.mx/explore/dataset/incidentes-viales-c5/table/?disjunctive.incidente_c4

```python
# This part of the code is used to read the files in the GitHub folder files, unzip the content and generate
# the file or dataframe
import pandas as pd
import glob
import os

def getDataFile(saving):
    os.system('unzip ./files/dataSet_incidentes-viales-c5.zip')
    files=glob.glob(os.path.join('dataSet_incidentes-viales-c5','*.csv'))

    data = pd.DataFrame()
    for x in files:
        info = pd.read_csv(x)
        data = pd.concat([data,info])

    if saving == 1:
        data.to_csv('./files/incidentes-viales-c5.csv',index=False,encoding='utf-8')
        os.system('rm -r ./dataSet_incidentes-viales-c5')
    return data

files = glob.glob('./files/*')
if not './files/incidentes-viales-c5.csv' in files:
    data = getDataFile(0)
else:
    filePath = './files/incidentes-viales-c5.csv'
    data = pd.read_csv(filePath)
```

### Cleaning the database
Cleaning the database should include finding all the blanks in the database as well as misspelling words or typos. This database does not present misspelling or typos in the columns. However, it has lot of white spaces and blanks.

In the code below, it is presenting the information about the null spaces by columns. The columns titled *delegacion_inicio*,*delegacion_cierre*,*latitud*,*longitud* and *geopoint* have null information. For that reason, it is needed to be corrected.

```python
data.info()
```

**The following part of the code also find the columns with null values**

```python
# Proccess to check all columns with blank elements
def columnListWithBlank(data):
    lista = data.columns.tolist()
    checkList = list()
    for i in lista:
        if len(data[i].isnull().unique()) > 1:
            checkList.append(i)
    return checkList

columnListWithBlank(data)
```

**Spliting the information to get the real_cases**
When readin the database, the reader should pay attention to a column named *codigo_cierre*, this column is very important because it has the information for the real cases. The code below helps to split the column by code and description:

```python
# Split codigo_cierre column in two columns
splitValues = data['codigo_cierre'].str.split(' ', n=1,expand=True)
splitValues[0] = splitValues[0].str.replace('(','')
splitValues[0] = splitValues[0].str.replace(')','')
data['codigo_cierre_val'] = splitValues[0]
data['codigo_cierre_desc'] = splitValues[1]
```

Once the column is splitted, it is needed to set the real cases. This will help to clean the database and also extract the crime information. Please refer to the database section and check the **Closing code** table.

```python
# Mark real cases as True and not real case as False
data.loc[(data['codigo_cierre_val'].isin(['A','I'])),'real_case'] = True
data.loc[(~data['codigo_cierre_val'].isin(['A','I'])),'real_case'] = False
```

Now, the table has the real cases that will help to work with crime information only. On the other hand, as it is mentioned before, there are some columns with null values that needs to be cleaned.

```python
# Check the delegacion_cierre information to retrieve parameter missing
# ej. check parameters in Coyoaccan and retrieve latitud
def validateParameters(parameter):
    delegaciones = data['delegacion_cierre'].drop_duplicates().tolist()
    for delegacion in delegaciones:
        data.loc[(data[parameter].isnull()) & (data['delegacion_cierre'].isin([delegacion])),parameter] = data.loc[(data['delegacion_cierre'].isin([delegacion])),parameter].describe().values[1]
    return data

# Process to eliminate the blank spaces into delegacion_cierre and delegacion_inicio
data.loc[(data['delegacion_inicio'].isnull()),'delegacion_inicio'] = data['delegacion_inicio'].describe().values[2]
data.loc[(data['delegacion_cierre'].isnull()),'delegacion_cierre'] = data['delegacion_cierre'].describe().values[2]

# fill blank spaces with delegaccion_cierre information
data = validateParameters('latitud')
data = validateParameters('longitud')

# calculate the geopoint based on the latitud and longitud
data['geopoint'] = data['latitud'].astype(str).str.cat(data['longitud'].astype(str), sep =",")
```

Finally, it is needed to check the time columns. To better work in the database it is neede to convert into datetime.

```python
# convert fecha_creacion and fecha_cierre to datetime
data['fecha_creacion'] = pd.to_datetime(data['fecha_creacion'])
data['fecha_cierre'] = pd.to_datetime(data['fecha_cierre'])
```

In the table below, you can check all the columns and check the data type as well as the null values. In this. case, the database is not presenting null values, and also it contains datetime types for the columns that use time.

```python
data.info()
```

### Selecting the information
The main problem in the Mexico-City is to identify where is the most conflictive and dangerous zones in order to increase the survillance and the seccurity. The following code help to identify the real cases reported as accidents and extract this information to be analysed further.

```python
# Selecting Accidental Cases
nonAccidents = data.loc[(data['real_case'].isin([False]))]
accidents = data.loc[(data['real_case'].isin([True]))]
accidents = accidents.reset_index(drop=True)
print("Accidents: ",'{:,.0f}'.format(accidents['folio'].count()),"\nNon accidents: ",'{:,.0f}'.format(nonAccidents['folio'].count()))
```

### Preprocesing using SciKit Learn libraries
The preprocesing module is one of the most important during the data analysis, because it help to recognize special paterns to make predictions, to classify the information and to easily identify close relations on the database.

The code below is used to preprocess the information and create a subgroups to easily analyse the database based on the seasonality, over the months and the time during the day. This labels help to create graphics that help to make decisions about the current system.

```python
from sklearn import preprocessing, linear_model, cluster
import seaborn as sns
import matplotlib.pyplot as plt

le_codcie = preprocessing.LabelEncoder()
le_codcie.fit(accidents['codigo_cierre'].values)
accidents['codigo_cierre_encode'] = le_codcie.transform(accidents['codigo_cierre'].values)

le_delini = preprocessing.LabelEncoder()
le_delini.fit(accidents['delegacion_inicio'].values)
accidents['delegacion_inicio_encode'] = le_delini.transform(accidents['delegacion_inicio'].values)

le_inc4 = preprocessing.LabelEncoder()
le_inc4.fit(accidents['incidente_c4'].values)
accidents['incidente_c4_encode'] = le_inc4.transform(accidents['incidente_c4'].values)

le_alarma = preprocessing.LabelEncoder()
le_alarma.fit(accidents['clas_con_f_alarma'].values)
accidents['clas_con_f_alarma_encode'] = le_alarma.transform(accidents['clas_con_f_alarma'].values)

le_tipo = preprocessing.LabelEncoder()
le_tipo.fit(accidents['tipo_entrada'].values)
accidents['tipo_entrada_encode'] = le_tipo.transform(accidents['tipo_entrada'].values)

le_delcie = preprocessing.LabelEncoder()
le_delcie.fit(accidents['delegacion_cierre'].values)
accidents['delegacion_cierre_encode'] = le_delini.transform(accidents['delegacion_cierre'].values)

le_mescie = preprocessing.LabelEncoder()
le_mescie.fit(accidents['mes_cierre'].values)
accidents['mes_cierre_encode'] = le_mescie.transform(accidents['mes_cierre'].values) + 1
```

Due to this situation, the traffic and the number of people in the city decreases considerably and the crime grows lot.


The other months correspond to Halloween, Day of the death in Mexico and also there are a specal holiday period of no more than 4 days. Also during this period, the people use to go to the beach or visit the graveyard during the day and night. This increases the number of thefts as well as pickpocketing. ### Weekly analysis
In the previous section, the analysis take place in a monthly basis. Nevertheless, it is also neede to check the daily basis to determine when is the most conflictive day of the week. In this case, the information is constructed based on the date *fecha_creacion* and generate a special label to easily identify the days in the graphic.

```python
accidents['creacion_day_cde'] = accidents['fecha_creacion'].apply(lambda x : x.dayofweek)
accidents.loc[(accidents['creacion_day_cde']==0),'creacion_day_name'] = 'Mon'
accidents.loc[(accidents['creacion_day_cde']==1),'creacion_day_name'] = 'Tue'
accidents.loc[(accidents['creacion_day_cde']==2),'creacion_day_name'] = 'Wed'
accidents.loc[(accidents['creacion_day_cde']==3),'creacion_day_name'] = 'Thu'
accidents.loc[(accidents['creacion_day_cde']==4),'creacion_day_name'] = 'Fri'
accidents.loc[(accidents['creacion_day_cde']==5),'creacion_day_name'] = 'Sat'
accidents.loc[(accidents['creacion_day_cde']==6),'creacion_day_name'] = 'Sun'
group_dayname_creacion = accidents.groupby(['creacion_day_name','año_cierre'])['folio'].count().reset_index().groupby('creacion_day_name')['folio'].mean().reset_index()
group_dayname_creacion['creacion_day_name'] = pd.Categorical(group_dayname_creacion['creacion_day_name'],categories=['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
group_dayname_creacion = group_dayname_creacion.sort_values(by=(['creacion_day_name']))
[fig,ax]=plt.subplots(figsize=(10,5))
sns.set(font_scale=3)
sns.lineplot(data=group_dayname_creacion.sort_values(by='creacion_day_name'), x='creacion_day_name', y='folio')
```

Analysing the graphic above, Friday is the most dangerous day of the week. It can easily explain because most of the people use to go the the pubs to have a drink, they usually go to the cinema and spend lot of people outside. This situation increases the number of thefts as well as pickpoketing. The following code expressed the percentage by day to compare with the graphic.

```python
result = accidents.groupby(['creacion_day_cde'],as_index=False)['folio'].count()/accidents['folio'].count()*100
result = result.rename(columns={'folio':'percent'})
result = result.reset_index()
result.loc[(result['index']==0),'creacion_day_name'] = 'Mon'
result.loc[(result['index']==1),'creacion_day_name'] = 'Tue'
result.loc[(result['index']==2),'creacion_day_name'] = 'Wed'
result.loc[(result['index']==3),'creacion_day_name'] = 'Thu'
result.loc[(result['index']==4),'creacion_day_name'] = 'Fri'
result.loc[(result['index']==5),'creacion_day_name'] = 'Sat'
result.loc[(result['index']==6),'creacion_day_name'] = 'Sun'
result = result.drop(['creacion_day_cde','index'],axis=1)
result = result.rename(columns={'creacion_day_name':'day_of_week'})
result[['day_of_week','percent']]
```

### Seasonality anlysis
In this work, a seasonality analysis takes place to validate when are the common holidays in the country as well as to determine if the accidents increases due to this date. To do that the following code determine the holidays based on the datetime. This holidays correspond to Mexican Calendar.

```python
import holidays
mx_holidays = holidays.Mexico()
accidents['is_holiday'] = accidents['fecha_creacion'].apply(lambda x: mx_holidays.get(x))
accidents.head()

le_ho = preprocessing.LabelEncoder()
accidents.loc[~accidents.is_holiday.isnull(),'is_holiday_encoded'] = le_ho.fit_transform(accidents[~accidents.is_holiday.isnull()].is_holiday)
```

```python
[fig,ax]=plt.subplots(figsize=(20,10))
sns.set(font_scale=3)
sns.barplot(data=accidents.groupby(['is_holiday','año_cierre']).folio.count().reset_index().groupby('is_holiday').folio.mean().reset_index(),
            x='folio',y='is_holiday')
```

The graphic above illustrates the most common holidays in Mexico, in this graphic the large number of accidents are presented during the Benito Juarez's birthday. This date commonly are related to the holy week and in some cases mexican people have extra days to relax.

Finally, we design a statsmodel to easily understand the number of accidents by day uring the last years.
The. following code, creates a model based on the 365 days of the year and generate the statistical variables to be used in the graphics.

```python
import statsmodels.api as sm
obj = sm.tsa.seasonal_decompose(accidents.groupby('fecha_creacion')['folio'].count(), period=365)

_,ax = plt.subplots(2,2,figsize=(20,10), sharex=True, squeeze=False)
sns.set(font_scale=1)
sns.lineplot(data=obj.observed.reset_index(), x='fecha_creacion', y='folio', ax=ax[0,0])
sns.lineplot(data=obj.resid.reset_index(), x='fecha_creacion', y='resid', ax=ax[0,1])
sns.lineplot(data=obj.seasonal.reset_index(), x='fecha_creacion', y='seasonal', ax=ax[1,0])
sns.lineplot(data=obj.trend.reset_index(), x='fecha_creacion', y='trend', ax=ax[1,1])

ax[0,0].set_title('Observed')
ax[0,1].set_title('Residual')
ax[1,0].set_title('Seasonal')
ax[1,1].set_title('Trend')

sns.set(font_scale=2)
```

In the graphics above, we can easily check the representation of all the accidents during the last years. Take a little time to see the trend graphic that show a decrement in the accidents in the city. This behaviour are presented in this way because it is showing how a decrease due to an increment of cameras in strategic position over the city.

```python
#accidents.groupby(['mes','mes_cierre','año_cierre'])['folio'].count().reset_index().groupby(['mes','mes_cierre'])['folio'].mean().reset_index()
result = accidents.groupby(['mes','año_cierre','delegacion_cierre'])['folio'].count().reset_index().groupby(['delegacion_cierre'])['folio'].mean().reset_index()
result = result.sort_values(by=['folio'],ascending=False)
```
