## Introduction
Mexico-City is considered as one of the most problematic and conflict cities in the world. Everyday, it is reported lot of accidents. These accidents most of the time weren't serious and sometimes were false alarms. The mexican government deccide to implement a reporting system to store all the incidents. This reporting system is know as C4 and consists in many different phases, for example lot of cameras in strategic points in the city, as well as problematic locations. To improve the system, the government decide to include help-buttons in the electric poles to help people report incidents. Due to this system, many people begin to use this controls to report an accident and many other people use this system to joke causing false-alarms.

### Objectives
* Check the database for unconsistencies
* Determine the most conflict locations
* Check the seasonality on the database
* Analyse the accidents to determine the timeperiod when those occurs
